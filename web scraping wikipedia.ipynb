```python
def state_scrapper(i):
    html=urlopen(f'https://en.wikipedia.org/wiki/{i}')
    bs=BeautifulSoup(html,'html.parser')
    
    
    formation_date=[i.find('td',class_='infobox-data').text  for i in bs.find_all("tr") if i.find('th',class_='infobox-label')!=None and 'Formation' in i.find('th',class_='infobox-label').text]
    
    
    capital=[i.find('td',class_='infobox-data').text  for i in bs.find_all("tr") if i.find('th',class_='infobox-label')!=None and 'Capital' in i.find('th',class_='infobox-label').text]
    
    
    area=[i.find('td',class_='infobox-data').text  for i in bs.find_all("tr") if i.find('th',class_='infobox-label')!=None and 'Total' in i.find('th',class_='infobox-label').text]
    area=" ".join(area)
    area=re.findall(r"^[\d,]+",area)

    
    population=[i.find('td',class_='infobox-data').text  for i in bs.find_all("tr") if i.find('th',class_='infobox-label')!=None and 'Total' in i.find('th',class_='infobox-label').text]
    population=''.join(population)
    
    population=re.findall(r"(?<=)",population)
    
    literacy=[i.find('td',class_='infobox-data').text  for i in bs.find_all("tr") if i.find('th',class_='infobox-label')!=None and 'Literacy' in i.find('th',class_='infobox-label').text]
    
    
    GDP=[i.find('td',class_='infobox-data').text  for i in bs.find_all("tr") if i.find('th',class_='infobox-label')!=None and 'Total' in i.find('th',class_='infobox-label').text]
    GDP=" ".join(GDP)
    GDP=re.findall(r"₹[\d.]+ \w+",GDP)
    
    
    
    population=[i.find('td',class_='infobox-data').text  for i in bs.find_all("tr") if i.find('th',class_='infobox-label')!=None and 'Total' in i.find('th',class_='infobox-label').text]
    population=''.join(population)
    population=re.findall(r'(?<=\)) [\d,]+(?=[₹\s])',population)
    
    
    
    HDI=[i.find('td',class_='infobox-data').text  for i in bs.find_all("tr") if i.find('th',class_='infobox-label')!=None and 'HDI' in i.find('th',class_='infobox-label').text]
    
    
    row=[i,formation_date,capital,area,population,literacy,GDP,HDI]
    
    return row
    
    
    
```

```python
STATES=pd.DataFrame()
for i in new_states:
    STATES=pd.concat([STATES,pd.DataFrame(state_scrapper(i))],axis=1)
    print(i)
    
STATES=STATES.T
```

```python
def returner(i):
    try:
        return i[0]
    except:
        return np.nan
for i in range(1,8):
    STATES[i]=STATES[i].apply(returner)
    
```

```python
STATES.rename(columns={0:'STATE/Union Territory',1:'Formation date',2:'capital',3:'area (sq km2)',4:'Population',5:'literacy_rate',6:'GDP',7:'HDI(human development index)'},inplace=True)
```
